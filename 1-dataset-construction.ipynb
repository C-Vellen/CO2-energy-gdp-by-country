# 2/ Drop columns and clean data (Nans and non-country lines)

```python
# keep intersting columns regarding Kaya Identity and drop the others:
kaya_columns = ['country', 'year', 'iso_code', 'population', 'gdp', 'co2', 
       'co2_per_capita', 'co2_per_gdp', 'co2_per_unit_energy', 'energy_per_capita',
       'energy_per_gdp', ]
dataset = dataset[kaya_columns]

# drop non-country lines :
# dataset["country"].unique()
non_countries = ['Africa', 'Africa (GCP)','Antarctica','Asia',
       'Asia (GCP)', 'Asia (excl. China and India)', 
       'Central America (GCP)', 'Europe', 'Europe (GCP)',
       'Europe (excl. EU-27)', 'Europe (excl. EU-28)',
       'European Union (27)', 'European Union (28)', 
       'High-income countries', 'International aviation',
       'International shipping', 'Kuwaiti Oil Fires',
       'Kuwaiti Oil Fires (GCP)',
       'Least developed countries (Jones et al.)', 
       'Low-income countries', 'Lower-middle-income countries', 'Middle East (GCP)',
       'Non-OECD (GCP)', 'North America',
       'North America (GCP)', 'North America (excl. USA)', 'OECD (GCP)', 'OECD (Jones et al.)',
       'Oceania', 'Oceania (GCP)', 'Ryukyu Islands (GCP)', 'South America',
       'South America (GCP)', 'Upper-middle-income countries',
       'World']
dataset = dataset[~dataset["country"].isin(non_countries)]
```

```python
kaya_columns.remove('iso_code') # we don't need iso_code for our analysis
dataset = dataset.dropna(subset=kaya_columns, ignore_index=True)
sns.heatmap(dataset.isna())
```

# 3/ Feature engineering

### The goal is to make all the terms of Kaya's identity appear in the dataset : 
<br><br>
$$
CO_{2}[kg] = \frac{CO_{2}[kg]}{energy[kWh]} * \frac{energy[kWh]}{gdp[\$]} * \frac{gdp[\$]}{population} * population
$$

with : 
- $CO_{2}[Mt]$ : Annual total CO2 emission from burning fossil fuels and industrial processes. Includes : transport, energy production, heating. Excludes : land-use change (impact of deforestation).
- $energy [kWh]$ : primary energy consumption per year
- $gdp [\$]$ : gross domestic product per year, in 2011$ prices
- $population$ : population by country
<br><br>
dataset : https://github.com/owid/co2-data
<br><br>
##### _see also data/owid-co2-notebook.csv for columns informations_
&rarr; **available data and corresponding columns of dataset :**
- $CO_{2}[Mt]$ : "co2"
<br><br>
- $ \frac{CO_{2}[kg]}{energy[kWh]} $ : "co2_per_unit_energy" = how many CO2 kg is emitted when 1kWh energy is consumed (high with coal, oil, low with solar, wind, nuclear)
<br><br>
- $ \frac{energy[kWh]}{gdp[\$]} $ : "energy_per_gdp" = how many energy ($) is necessary to produce wealth ($) (high : unefficient, poor isolation, low: efficient, good yield)
<br><br>
- $ population $ : "population"

&rarr; **data to build :**
- $ energy [kWh] $
<br><br>
- $ \frac{gdp[\$]}{population} $ : add column "gdp_per_capita" by dividing "gdp" per "population"

&rarr; **for verifications, we add temporary columns :**
- "co2_kaya" : CO2 emissions calculated using the Kaya Identity method
- "co2_kaya_diff" : Difference in CO2 emissions calculated using the Kaya Identity method, should be ~ 0
- "co2_kaya_diff_prct" : Percentage difference in CO2 emissions calculated using the Kaya Identity method, should be ~ 0%
<br><br>
- "energyi_vs_energyj_diff" : Difference in energy estimation using 3 calculation methods (i,j = 1, 2, 3), should be ~ 0
- "energyi_vs_energyj_diff_prct" : Percentage difference in energy estimation using 3 calculation methods, should be ~ 0%



### 3.1/ add column : $ \frac{gdp[\$]}{population} $

```python
df_kaya = dataset.copy()
# add column "gdp_per_capita"
df_kaya.loc[:,'gdp_per_capita'] = dataset["gdp"]/dataset["population"]
```

```python
# NaN values :
sns.heatmap(df_kaya.isna())
```

```python
df_kaya.shape
```

##### Check dataset values are consistent : co2_kaya = co2 

```python
# add verification columns to compare 'co2' and 'co2' calculated with Kaya Identity
df_verif = df_kaya['co2_per_unit_energy'] * df_kaya['energy_per_gdp'] * df_kaya['gdp_per_capita'] * df_kaya['population']
df_kaya.loc[:,'co2_kaya'] = df_verif/1e9  # convert to million tonnes
df_kaya.loc[:,'co2_kaya_diff'] = df_kaya['co2'] - df_kaya['co2_kaya']
df_kaya.loc[:,'co2_kaya_diff_prct'] = df_kaya['co2_kaya_diff'] / df_kaya['co2'] * 100

# verification co2_kaya = co2 
df_kaya[['co2_kaya_diff_prct']].describe()
```

```python
# verification co2_kaya = co2 visualisation
sns.boxplot(x=df_kaya['co2_kaya_diff_prct'])
```

```python
# inconsistent values (more than 1% difference):
df_kaya_inconsistent_values = df_kaya.query("co2_kaya_diff_prct < -1 or co2_kaya_diff_prct > 1")
df_kaya_inconsistent_values[['country', 'year', 'population', 'co2', 'co2_kaya', 'co2_kaya_diff_prct']]
```

##### CONCLUSION : the calculated co2 with Kaya identity is consistent with the co2 values from the dataset.
- 7745 values out of 7757 ( 99.8%) have a difference of less than 1% between co2 and co2_kaya.
- 12 values have a difference of more than 1% and will be excluded from further analysis.
The discrepancies may be due to rounding errors or data inconsistencies in the original dataset.

```python
df_kaya = df_kaya.query("co2_kaya_diff_prct >= -1 and co2_kaya_diff_prct <= 1")
df_kaya.drop(columns=['co2_kaya', 'co2_kaya_diff', 'co2_kaya_diff_prct'], inplace=True)    
df_kaya.shape
```

### 3.2/ add column "energy"

```python
# add column "energy" from 3 calculation methods
df_energy1 = df_kaya['energy_per_gdp'] * df_kaya['gdp']
df_kaya.loc[:,'energy1'] = df_energy1
df_energy2 = df_kaya['energy_per_capita'] * df_kaya['population']
df_kaya.loc[:,'energy2'] = df_energy2
df_energy3 = df_kaya['co2'] / df_kaya['co2_per_unit_energy'] * 1e9
df_kaya.loc[:,'energy3'] = df_energy3     

# verify that the 3 energy calculation methods give the same result
df_kaya.loc[:,'energy1_vs_energy2_diff'] = df_kaya['energy1'] - df_kaya['energy2']
df_kaya.loc[:,'energy1_vs_energy3_diff'] = df_kaya['energy1'] - df_kaya['energy3']
df_kaya.loc[:,'energy2_vs_energy3_diff'] = df_kaya['energy2'] - df_kaya['energy3']  
df_kaya.loc[:,'energy1_vs_energy2_diff_prct'] = df_kaya['energy1_vs_energy2_diff'] / df_kaya['energy1'] * 100
df_kaya.loc[:,'energy1_vs_energy3_diff_prct'] = df_kaya['energy1_vs_energy3_diff'] / df_kaya['energy1'] * 100          
df_kaya.loc[:,'energy2_vs_energy3_diff_prct'] = df_kaya['energy2_vs_energy3_diff'] / df_kaya['energy2'] * 100
df_kaya[['energy1_vs_energy2_diff_prct', 'energy1_vs_energy3_diff_prct', 'energy2_vs_energy3_diff_prct']].describe()   

```

```python
# visualize the differences between the 3 energy calculation methods
fig, ax = plt.subplots(1, 3, figsize=(8,2))
fig.suptitle("Comparaison des mÃ©thodes de calcul de l'Ã©nergie")
for i,v in enumerate(['energy1_vs_energy2_diff_prct', 'energy1_vs_energy3_diff_prct', 'energy2_vs_energy3_diff_prct']):
    sns.boxplot(x=df_kaya[v], ax=ax[i])
    ax[i].set_xlabel(v)
plt.tight_layout()
plt.show()
```

```python
# display inconsistent energy calculation values
for i,v in enumerate(['energy1_vs_energy2_diff_prct', 'energy1_vs_energy3_diff_prct', 'energy2_vs_energy3_diff_prct']):
    print('_'*20)
    print(v)
    print(df_kaya.query(f"{v} < -1 or {v} > 1")[["country", "year", "energy1", "energy2", "energy3"]])
```

##### Conclusion : 
energy1 & energy2 are very consistent (difference < 0.6%), energy3 is consistent (only 1 value with some difference >1% but < 1.6%)

```python
df_kaya["energy"] = (df_kaya["energy1"] + df_kaya["energy2"]) / 2
```
